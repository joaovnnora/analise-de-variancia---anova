# Análise de variância

### 1. Definições e tabela ANOVA

Para fazermos a comparação de médias em três ou mais populações definidas, consideremos um modelo estátisco em que cada observação $Y$ pode ser decomposta em duas componentes: *sistemática* e *aleatória*.

Matematicamente, podemos escrever 
$$
Y = \mu + e
$$

Assim, se $Y$ resenta a observação associada a uma unidade experimental, a parte sistemática $\mu$ pode ser vista como a média populacional, que é fixa, e a parte aleatória $e$ como informação que podem influir nas observações mas não são incorporadas em $\mu$.

Suponha que estamos interressados em comparar as médias de $K$ populações, isto é, queremos testar
$$
H_{0} : \mu_{1} = \mu_{2} = \dots = \mu_{K} ; \\
H_{a} : \text{pelo menos uma das médias } \mu_{i} \text{ é diferente das demais.}
$$

Para tanto, obtemos $K$ amostras independentes, com $m$ indivíduos cada uma delas. Nesta situações, o modelo estatístico para $j$-ésima unidade experimental, da $i$-ésima populção é dado por
$$
\text{Modelo 1: } \ Y_{ij} = \mu_{i} + e_{ij}, \quad i = 1, \dots, K; \quad j = 1, \dots, m.
$$

Caso a hipótese $H_{0}$ seja verdadeira, teremos que todas as médias para as $K$ populações serão iguais a um valor comum $\mu$. O modelo pode então ser descrito como
$$
\text{Modelo 0: } \ Y_{ij} = \mu + e_{ij}^{*}, \quad i = 1, \dots, K; \quad j = 1, \dots, m.
$$

Uma forma de levar em conta a informaçõa não explicada pela forma sistemática é através das somas de quadrados
$$
\sum_{i=1}^{K} \sum_{j=1}^{m} e_{ij}² = \sum_{i=1}^{K} \sum_{j=1}^{m} (Y_{ij} - \mu_{i})^2 \qquad (1) 
\quad \text{e} 
$$
$$
\qquad \sum_{i=1}^{K} \sum_{j=1}^{m} (e_{ij}^*)² = \sum_{i=1}^{K} \sum_{j=1}^{m} (Y_{ij} - \mu)^2. \qquad (2)
$$

As expressões envolvem medidas $\mu$ e $\mu_{i}$ desconhecidas. Então, utilizaremos os dados para obter as etimativas correspondentes. Como no Modelo 1 estamos considerando diferentes valores de $\mu_{i}$ para as diferentes $K$ populações, consideramos os dados oriundos de cada uma das populações para estimar a correspondente média. Assim, segundo o Modelo 1 temos
$$
\hat{\mu_{i}} = \frac{1}{m} \sum_{j=1}^m Y_{ij} = \bar{Y_{i}}, \ i = 1, \dots , K. \qquad (3)
$$

Para o Modelo 0, como assumimos que todas as populações tem o mesmo $\mu_{i}$, utilizamos o estimador
$$
\hat{\mu} = \frac{1}{Km} \sum_{i=1}^{K} \sum_{j=1}^m Y_{ij} = \bar{Y} \qquad (4)
$$

Substituindo as estimativas nas correspondentes somas de quadrados, definimos as quantidades *soma de quadrados dentro* (**SQD**) e *soma de quadrados total* (**SQT**) da seguinte forma:
$$
\text{SQD} = \sum_{i=1}^{K} \sum_{j=1}^m (Y_{ij} - \hat{\mu_{i})}^2 = \sum_{i=1}^{K} \sum_{j=1}^m (Y_{ij} - \bar{Y_{i}})² = \sum_{i=1}^{K} \sum_{j=1}^m Y_{ij}² - m \sum_{i=1}^{K} \bar{Y_{i}²} \quad \text{(Rejeitando } H_{0} \text{)};
$$

$$
\text{SQT} =  \sum_{i=1}^{K} \sum_{j=1}^m (Y_{ij} - \hat{\mu})^2 =  \sum_{i=1}^{K} \sum_{j=1}^m (Y_{ij} - \bar{Y_{i}})^2 =  \sum_{i=1}^{K} \sum_{j=1}^m Y_{ij}² - mK\bar{Y²} \quad \text{(Não rejeitando } H_{0} \text{)} . 
$$

A diferença entre SQT e SQD representa *soma de quadrados entre* e será denotada por SQE, isto é,
$$
SQE = SQT - SQD.
$$

Das expressões de SQD e SQT segue que:
$$
SQE = (\sum_{i=1}^{K} \sum_{j=1}^m Y_{ij}² - mK\bar{Y²} ) - (\sum_{i=1}^{K} \sum_{j=1}^m Y_{ij}² - m \sum_{i=1}^{K} \bar{Y_{i}²}) = - mK\bar{Y²} + m \sum_{i=1}^{K} \bar{Y_{i}²} = m( \sum_{i=1}^{K} \bar{Y_{i}²} - K\bar{Y²} ).
$$

Levando em consideração que cada uma das somas de qadrados envolve um certo número de quantidades que estão sendo *estimadas* e o número de observações nas amostras, definimos os correspondentes quadrados médios:
$$
QMT = \frac{SQT}{Km - 1}: \text{quadrado médio total};
$$

$$
QMD = \frac{SQD}{Km - K} = \frac{SQD}{K(m - 1)}: \text{quadrado médio dentro};
$$

$$
QME = \frac{SQE}{K - 1} : \text{quadrado médio entre}.
$$

O teste estatístico para a hipótese $ H_{0}$ envolve os quadrados médios. Se a hipótese $H_{0}$ não for verdadeira, então, o Modelo 1 deve ser mais adequado aos dados que o Modelo 0. Em outras palavras, os resíduos produzidos pelo Modelo 1 serão menores que os do Modelo 0. Analisando sob este ângulo, podemos interpretar QME como sendo a informação contida nos dados que é captada pelo Modelo 1, enquanto o QMD representa a informação não explicada pelo Modelo 1. Portando, se o QME for grande comparado à QMD, a parte sistemática do Modelo 1 estará captando grande prte da informação dos dados e a hipótese $H_{0}$ deverá ser rejeitada. Definimos então
$$
F = \frac{QME}{QMD}.
$$

Assim, quanto maior $F$ maior evidência contra $H_{0}$.

Para caracterizarmos o valor crítico a partir do qual rejeitamos $H_{0}$, precisamos encontrar a distribuição de probabilidade para $F$. Supondo as seguintes condições:
- $Y_{ij}$ são variáveis aleatórias independentes;
- $Y_{ij}$ é Normal com médio $\mu_{i}$, $i$ = 1, $\dots$, $K$ e $j$ = 1, $\dots$, $m$;
- tdas as $K$ populações tem variâncias iguais a $\sigma^2$,

pode ser mostrado que
$$
F \sim F(K - 1, K(m - 1))
$$

isto é, a quantidade $F$ tem distribuição de Fisher-Snedecor com $K$ - 1 e K(m - 1) graus de liberdade. Temos, agora, condições de encontrar o valor crítico de $f_{c}$ e determinar a região crítica do teste, que será da forma
$$
RC = \{f\in \mathbb{R⁺}: f > f_{c}\}.
$$

A discussão é resumida na tabela a seguir.
$$
\begin{array}{|c|c|c|c|c|}
\hline
\text{Fonte de variação} & \text{Graus de liberdade} & \text{Soma de quadrados} & \text{Quadrado médio} & \text{F} \\
\hline
\text{Entre} & K - 1 & SQE & QME & \frac{QME}{QMD} \\
\hline
\text{Dentro} & K(m - 1) & SQD & QMD & \\
\hline
\text{Total} & Km - 1 & SQT & & \\
\hline
\end{array}
$$

### 2. Exemplo

A gerência de um depósito que armazena cargas aéreas de pequeno porte está estudando o peso das cargas que chegam ao seu terminal no interior de São Paulo. Usualmente, o terminal recebe 4 tipos de cargas: doméstica (D), administrativa (A), equipamentos industriais (E) e outros tipos (O). Deseja-se verificar se, em média, existem diferenças entre os pesos dos 4 tipos de cargas. Ao longo de 1 mês, cargas foram colhidas aleatoriamente e seus pesos foram aferidos, fornecendo os dados (em kg).

Começamos importando as bibliotecas necessárias para a análise.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

Então, fazemos a leitura e visualizamos os dados e vendo a análise descritiva deles.

```python
data = pd.read_csv('../data/dados.csv', sep=';')

data
```

```python
mediaD = data['D'].mean()
medianaD = data['D'].median()
sigmaD = data['D'].std()

mediaA = data['A'].mean()
medianaA = data['A'].median()
sigmaA = data['A'].std()

mediaE = data['E'].mean()
medianaE = data['E'].median()
sigmaE = data['E'].std()

mediaO = data['O'].mean()
medianaO = data['O'].median()
sigmaO = data['O'].std()

descritivo = {
    'Medidas descritivas' : ['Media', 'Mediana', 'Desvio Padrão'],
    'D' : [mediaD, medianaD, sigmaD],
    'A' : [mediaA, medianaA, sigmaA],
    'E' : [mediaE, medianaE, sigmaE],
    'O' : [mediaO, medianaO, sigmaO]
}

descritivo = pd.DataFrame(descritivo)

descritivo
```

As medidas descritivas e os próprios dados nos sugerem algumas coisas. As cargas do tipo E parecem tender a ter maior peso e as cargas do tipo D parecem tender a ter menor peso. Também, percebemos que os desvios padrões encontram-se razoavelmente próximos.

Ainda fazendo análise exploratória, vamos imprimir os histogramas e boxplots de cada carga.

```python
f, ax_hist = plt.subplots(2,2,figsize=(12,8))

sns.histplot(data = data, x = "D", ax = ax_hist[0][0], kde=True)
ax_hist[0][0].axvline(mediaD, color='r', linestyle='--', label="Média")
ax_hist[0][0].axvline(medianaD, color='g', linestyle='-', label="Mediana")
ax_hist[0][0].legend()

sns.histplot(data = data, x = "A", ax = ax_hist[0][1], kde=True)
ax_hist[0][1].axvline(mediaA, color='r', linestyle='--', label="Média")
ax_hist[0][1].axvline(medianaA, color='g', linestyle='-', label="Mediana")
ax_hist[0][1].legend()

sns.histplot(data = data, x = "E", ax = ax_hist[1][0], kde=True)
ax_hist[1][0].axvline(mediaE, color='r', linestyle='--', label="Média")
ax_hist[1][0].axvline(medianaE, color='g', linestyle='-', label="Mediana")
ax_hist[1][0].legend()

sns.histplot(data = data, x = "O", ax = ax_hist[1][1], kde=True)
ax_hist[1][1].axvline(mediaO, color='r', linestyle='--', label="Média")
ax_hist[1][1].axvline(medianaO, color='g', linestyle='-', label="Mediana")
ax_hist[1][1].legend()
```

```python
sns.boxplot(data = data)
plt.ylabel("Pesos das cargas")
plt.title("Boxplot")
plt.show()
```

A análise dos histogramas e dos boxplots dão ainda mais força para nossas suposições.

Antes de usar ANOVA, vamos verificar a normalidade dos dados usando o teste de Shapiro.

```python
shapiro_test, pD = stats.shapiro(data['D'])
shapiro_test, pA = stats.shapiro(data['A'])
shapiro_test, pE = stats.shapiro(data['E'])
shapiro_test, pO = stats.shapiro(data['O'])

print(pD, pA, pE, pO)
```

Os p-valores dos dados nos indicam a normalidade deles.

Agora, como os dados satisfazem nossas três condições, podemos usar ANOVA.

Nesse exemplo, temos K = 4 grupos e m = 7 observações por grupo. Agora, vamos conseguir os outros dados para construir a tabela.

Primeiro, vamos adequar os dados para podermos usar o pacote do statsmodel.

```python
dataMelt = pd.melt(data.reset_index(), id_vars=['index'], value_vars=['D', 'A', 'E', 'O'])
```

Agora, calculamos e construimos a tabela ANOVA.

```python
K = 4
m = 7

model = ols('value ~ C(variable)', data=dataMelt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

SQE = anova_table['sum_sq']['C(variable)']
SQD = anova_table['sum_sq']['Residual']
SQT = SQE + SQD
QME = SQE / (K - 1)
QMD = SQD / (K * (m - 1))
F = QME / QMD
total = (K * m) - 1
grausDeLiberdadeEntre = K - 1
grausDeLiberdadeDentro = K * (m - 1)

tabela = {
    'Fonte de variação' : ['Entre', 'Dentro', 'Total'],
    'Graus de liberdade' : [grausDeLiberdadeEntre, grausDeLiberdadeDentro, total],
    'Soma de quadrados' : [SQE, SQD, SQT],
    'Quadrado médio' : [QME, QMD, ''],
    'F' : [F, '', '']
}

tabela = pd.DataFrame(tabela)

tabela
```

Temos nosso $f_{obs}$ = 127.57.

Agora, através da distribuição F, com 3 e 24 graus de liberdade e, considerando $\mu$ = 5%, vamos obter $f_{c}$.

```python
alpha = 0.05

valor_critico_f = stats.f.ppf(1 - alpha, grausDeLiberdadeEntre, grausDeLiberdadeDentro)

valor_critico_f
```

Temos então $f_{c}$ = 3.008.

Portando, como calculamos nosso $f_{obs}$ = 127.57 > $f_{c}$, concluímos que, ao nível de significância de 5%, as médias de peso dos grupos são diferentes, assim, rejeitando $H_{0}$.
